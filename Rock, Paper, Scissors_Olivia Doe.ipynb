```python
#Create a program where the user can play a game of rock, paper, scissors against a computer. Set a play choice (rock, paper, or scissors) for the computer player and during the program, request for the user to input their play choice.
#
#The rules in Rock, Paper, Scissors are:
#
#    Rock beats scissors
#    Scissors beats paper
#    Paper beats rock
#
#If the user wins, then display "You win!" If they lose against the computer, then display "You lose." If the computer and the user have the same play choice, display "It's a draw."
#
#BONUS: Randomize the play choice for the computer (hint: Python has a library called random). Also ask the user if they would like to play again. If they do, start the game again but if not, thank the user for playing the game.
```

```python
import random


CHOICES = ("rock", "paper", "scissors")


#Check if user will beat computer. Returns true if user wins,
# returns false if user loses
def check_choice(user_action, comp_action):
    if ((user_action == "rock" and comp_action == "scissors")or
          (user_action == "scissors" and comp_action == "paper")or
          (user_action == "paper" and comp_action == "rock")):
         return True
    else:
        return False

    
#Generate a random choice from tuple of rock, paper, scissors    
def generate_comp_choice():
    return random.choice(CHOICES)

#Convert user choice from int to its equivalent string
def convert_user_choice(user_action):
    user_choice_dictionary = {1:"rock", 2:"paper", 3:"scissors"}
    return user_choice_dictionary[user_action]

```

```python
#main driver
while True: 
    user_choice = int(input("""Choose\n
    [1] rock\n
    [2] paper\n
    [3] scissors\n\n\t"""))
    #user choice converted to equivalent
    user_choice = convert_user_choice(user_choice)
    #generate computer choice
    comp_choice = generate_comp_choice()
    
    print("You picked:", user_choice, "\nComputer picked:", comp_choice)
    
    #check winner
    if check_choice(user_choice, comp_choice):
        message = "You win!"
    elif user_choice == comp_choice:
        message = "It's a draw"
    else:
        message = "You lose!"
        
    print("\n", message)
    keep_playing = input("Continue (y/n): ")
    keep_playing = keep_playing.lower()
    if keep_playing == "n":
        break
print("\nThank You for playing!")
```
